# exemple d'utilisation et principe de l'algorithme de dichotomie
Reprenons l'énoncé de l'exercice 4 su td1 (chapitre 4).

On considère pour tout $x$ de $[-10;10]$ la fonction $f(x)=x^3-3x^2+x+3$.

On démontre dans l'exercice que l'équation $f(x)=0$ admet une unique solution, $\alpha$, puis on demande un arrondi de $\alpha$ à $10^{-2}$ près.

Le principe est le suivant : 
on sait que $\alpha\in[-10;10]$, on calcule alors le centre de l'intervalle : $\displaystyle\frac{-10+10}{2}=0$

Soit $\alpha\in[-10;0]$, soit $\alpha\in[0;10]$.

On démontre assez facilement que $\alpha\in[-10;0]$, on calcule à nouveau le centre de l'intervalle : ... -5.


Soit $\alpha\in[-10;-5]$, soit $\alpha\in[-5;0]$.

On continue ainsi jusqu'à ce que l'intervalle ait une amplitude inférieure à la précision voulue : à chaque étape, l'amplitude de l'intervalle diminue car divisé par 2, cette amplitude tend donc vers 0.
